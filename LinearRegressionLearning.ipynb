```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Download latest version
path = kagglehub.dataset_download("simpleparadox/bostonhousingdataset")

print("Path to dataset files:", path)
```

```python
dtset = pd.read_csv("/kaggle/input/bostonhousingdataset/boston-housing-dataset.csv")
dtset.head(100)
print(dtset.shape)
print(dtset)
```

```python
data = dtset.loc[:, ['RAD', 'AGE']]
print(data.head(5))
```

```python
data.plot(x='RAD', y='AGE', style='o')
plt.xlabel('RAD')
plt.ylabel('AGE')
plt.show()
```

```python
import sklearn
import sklearn
from sklearn.linear_model import LinearRegression
X = data['RAD'].values.reshape(-1, 1)  # Reshape to 2D
y = data['AGE'].values  # y can stay 1D
regressor = LinearRegression()
regressor.fit(X, y)
data.plot(x = 'RAD', y = 'AGE', style='o', label='Data Points')
plt.plot(X, regressor.predict(X), color='Gold', linewidth=5, label='Regression Line')
plt.xlabel("RAD")
plt.ylabel('AGE')
plt.legend()
plt.show()
```
